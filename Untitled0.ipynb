```python
fashion_mnist = keras.datasets.fashion_mnist  # Loading the Fashion MNIST data
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
X_train_full.shape  # every image is represented as a 28 × 28 array
```

```python
X_train_full=np.reshape(X_train_full,(60000,784))
```

```python
X_test.shape
```

```python
X_test=np.reshape(X_test,(10000,784))
```

```python
X_train_full.dtype
```

```python
X_valid, X_train = X_train_full[:5000]/255.0, X_train_full[5000:]/255.0
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress",
               "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
y_train.size
```

```python
y_train
```

```python
one_hot_encode = []
for i in range(60000):
    arr = list(np.zeros(len(class_names), dtype = int))
    arr[y_train_full[i]] = 1
    one_hot_encode.append(arr)
one_hot_encode=np.array(one_hot_encode)
```

```python
one_hot_encode_test = []
for i in range(10000):
    arr = list(np.zeros(len(class_names), dtype = int))
    arr[y_test[i]] = 1
    one_hot_encode_test.append(arr)
one_hot_encode_test=np.array(one_hot_encode_test)
```

```python
one_hot_encode_valid, one_hot_encode_train = one_hot_encode[:5000], one_hot_encode[5000:]
```

```python
def viz_image(index):
    reshaped_image_i=np.reshape(X_train[index],(28,28))
    plt.imshow(reshaped_image_i, cmap='hot', interpolation='nearest')
    plt.axis('off')
    plt.savefig("test.png", bbox_inches='tight')
    plt.show()
```

```python
viz_image(0)
```

```python
input_ = keras.layers.Input(shape=(784,))
hidden1 = keras.layers.Dense(200, activation="relu", name = "hidden_layer_1")(input_)
hidden2 = keras.layers.Dense(100, activation="relu")(hidden1)
output = keras.layers.Dense(10, activation="sigmoid")(hidden2)
```

```python
model = keras.Model(inputs=[input_], outputs=[output])
```

```python
model.summary()
```

```python
sgd = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(loss="categorical_crossentropy", optimizer = sgd, metrics=["accuracy"])
```

```python
model.fit(X_train,one_hot_encode_train,epochs=10,batch_size=100,
          validation_data=(X_valid, one_hot_encode_valid))
```

```python
model.evaluate(X_test, one_hot_encode_test, batch_size=None)
```

```python
first_hidden_layer_model = tf.keras.Model(inputs=model.input, outputs=model.get_layer("hidden_layer_1").output)
hidden_layer_1_data = first_hidden_layer_model.predict(X_test)
tsne = TSNE(n_components=2, random_state=42)
tsne_embeddings = tsne.fit_transform(hidden_layer_1_data)
plt.figure(figsize=(8, 6))
plt.scatter(tsne_embeddings[:, 0], tsne_embeddings[:, 1], c=np.argmax(one_hot_encode_test, axis=1))
plt.colorbar()
plt.title("t-SNE Visualization of Hidden Layer 1 Data")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.show()
```

```python
umap_model = umap.UMAP(n_components=2)
umap_embeddings = umap_model.fit_transform(hidden_layer_1_data)

plt.figure(figsize=(8, 6))
plt.scatter(umap_embeddings[:, 0], umap_embeddings[:, 1], c=np.argmax(one_hot_encode_test, axis=1))
plt.colorbar()
plt.title("UMAP Visualization of Hidden Layer 1 Data")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()
```
